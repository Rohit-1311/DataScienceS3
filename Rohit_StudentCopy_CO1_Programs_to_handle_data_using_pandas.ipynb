```python
#2. Write a python program to Generate the series of dates from 1st September, 2023 to 25th September, 2023 (both inclusive).
import pandas as pd
series = pd.Series(pd.date_range('2023-09-01','2023-09-25',freq='D'))
print(series)
```

```python
# 5. Given is a dataframe showing name, occupation,salary of people. Find the average salary per occupation.
import pandas as pd
details = { "Name":['Rohit','Jissmon','Adarsh','Vishnu','Glodin','Dev'],
           "Occupation":['Manager','Manager','Programmer','Programmer','Analyst','Analyst'],
            "Salary" : [20000,25000,28000,27000,30000,35000]
            }

df = pd.DataFrame(details)
print(df)
o_avg_age = df.groupby('Occupation')['Salary'].mean()
print("\nThe average salary per occupation is: ")
print(o_avg_age)
```
